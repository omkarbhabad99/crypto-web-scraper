```python
import json
from requests import Session, exceptions

url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'  

parameters = {
    'start': '1',
    'limit': '100',  # Reduce limit if needed
    'convert': 'USD'
}

headers = {
    'Accepts': 'application/json',
    'X-CMC_PRO_API_KEY': '599442f0-55cc-4a7b-9324-ee22525d5c45',
}

session = Session()
session.headers.update(headers)

try:
    response = session.get(url, params=parameters)
    data = response.json()

    # Save the data to a file
    with open("crypto_data.json", "w") as file:
        json.dump(data, file, indent=4)

    print("Data saved to crypto_data.json")

except (exceptions.ConnectionError, exceptions.Timeout, exceptions.TooManyRedirects) as e:
    print(f"Error: {e}")
```
